Reading the file with numpy

```python

import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
chainsAngles = genfromtxt('../build/chainAngle.csv', delimiter=',')
```

```python
angles = chainsAngles[:,1]
```

Number of chains:

```python
angles.size
```

Drawing a histogram chart of chains' angle. Each angle value is the angle between each chain
and the bed (positive x-axis). The direction of a chain is assumed to be always +z. 
The chart shows how many chains are within each angle bin.
The peak is at 160-180 degrees where 25 chains are aligned in that range of angles.
The graph should be seen in a periodic way where 0 and 180 the same point. Becuase
 

```python

bins = 8
plt.hist(angles, bins=bins)
plt.ylabel('Chains Count')
plt.xlabel('Angle (Degree)')
plt.xticks(np.arange(0,181,step=180/bins));
plt.xlim(xmin=0,xmax=180);
```
